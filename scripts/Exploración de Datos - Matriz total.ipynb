```python
from pathlib import Path
import pandas as pd

# Ruta 
BASE = Path.cwd()
PATH_CCQ = BASE / "data_raw" / "mafi" / "matriz_total_2021_2025_01.xlsx"

# Lectura
df_matriz = pd.read_excel(PATH_CCQ, engine="openpyxl")

# Resumen
print(f"Filas: {len(df_matriz):,} | Columnas: {len(df_matriz.columns)}")
print("Columnas:")
print(list(df_matriz.columns))

display(df_matriz.head(10))
```

```python
# EDA Matriz Total 2021–2025 (Scoring de riesgo)
import pandas as pd, numpy as np, matplotlib.pyplot as plt, re
from matplotlib.ticker import PercentFormatter
from cycler import cycler
from pathlib import Path

# Estilo visual
plt.rcParams.update({
    "figure.dpi":110,"axes.spines.top":False,"axes.spines.right":False,"axes.grid":True,
    "grid.alpha":0.18,"axes.titleweight":"bold","axes.titlesize":12,"axes.labelsize":10,
    "xtick.labelsize":9,"ytick.labelsize":9,"legend.frameon":False,"legend.fontsize":9,
    "axes.prop_cycle":cycler(color=["#4B2E83","#005BBB","#7F56D9","#1F7A8C","#E67E22","#2D3748","#3BA99C"])
})
def _annot_pct(ax):
    for p in ax.patches:
        h=p.get_height()
        if not (np.isnan(h) or h==0):
            ax.annotate(f"{h:.1f}%",(p.get_x()+p.get_width()/2,h),
                        ha="center",va="bottom",fontsize=9,xytext=(0,3),textcoords="offset points")
def bar_pct(s, title, xlabel, top=None, rot=45, width=0.7):
    s=s.dropna()
    if s.empty: print(f"(sin datos para {title})"); return
    pct=(s.value_counts(normalize=True)*100).sort_values(ascending=False)
    if top: pct=pct.head(top)
    fig,ax=plt.subplots(figsize=(9.5,4)); pct.plot(kind="bar",ax=ax,width=width)
    ax.set_title(title); ax.set_xlabel(xlabel); ax.set_ylabel("% del total")
    ax.yaxis.set_major_formatter(PercentFormatter(100)); _annot_pct(ax)
    plt.xticks(rotation=rot,ha="right"); plt.tight_layout(); plt.show()
def hist_nice(x, title, xlabel, bins=30):
    x=x.dropna()
    if x.empty: print(f"(sin datos para {title})"); return
    fig,ax=plt.subplots(figsize=(9.5,4))
    ax.hist(x.values, bins=bins)
    ax.set_title(title); ax.set_xlabel(xlabel); ax.set_ylabel("frecuencia")
    plt.tight_layout(); plt.show()
def box_by_cat(x, g, title, xlabel, ylabel):
    df=pd.DataFrame({"x":x,"g":g}).dropna()
    if df.empty or df["g"].nunique()<2: 
        print(f"(sin datos suficientes para {title})"); return
    order=df["g"].value_counts().index.tolist()[:12]
    data=[df.loc[df["g"]==k,"x"].values for k in order if (df["g"]==k).sum()>0]
    if len(data)<2: print(f"(sin datos suficientes para {title})"); return
    fig,ax=plt.subplots(figsize=(10,4))
    bp=ax.boxplot(data, labels=[str(k) for k in order], showfliers=False, patch_artist=True)
    for patch,c in zip(bp["boxes"], ["#C8B6FF","#A78BFA","#7C3AED","#5B21B6","#4338CA","#1E3A8A","#4B2E83","#3BA99C","#005BBB","#E67E22","#2D3748","#1F7A8C"][:len(bp["boxes"])]):
        patch.set_facecolor(c); patch.set_alpha(0.65); patch.set_edgecolor("#444")
    ax.set_title(title); ax.set_xlabel(xlabel); ax.set_ylabel(ylabel)
    plt.xticks(rotation=45,ha="right"); plt.tight_layout(); plt.show()

# Normalizar encabezados y mapear sinónimos
def snake_simple(cols):
    out=[]
    for c in cols:
        c=str(c).strip()
        out.append(c)
    return out

df = df_matriz.copy()
df.columns = snake_simple(df.columns)

def find_col(patterns, cols):
    pats = patterns if isinstance(patterns,(list,tuple)) else [patterns]
    for p in pats:
        for c in cols:
            if re.search(p, str(c), flags=re.I):
                return c
    return None

col_valor_mora   = find_col([r"^valor\s*mora$"], df.columns) or find_col([r"\bvalor.*mora\b"], df.columns)
col_media_score  = find_col([r"^media\s*score$"], df.columns)
col_mora_text    = find_col([r"^mora$"], df.columns) 
col_cliente      = find_col([r"^cliente$"], df.columns) or find_col([r"identificaci[oó]n|id\b"], df.columns)
col_mes          = find_col([r"^mes$"], df.columns)
col_anio         = find_col([r"^a[nñ]o[b]?$","^anio$","^año$"], df.columns)
col_valor_liq    = find_col([r"^valor\s+a\s+liquidar$"], df.columns)
col_valor_fact   = find_col([r"^valor\s+factura$"], df.columns)
col_valor_fin    = find_col([r"^valor\s*financiado$"], df.columns)  
col_estado_final = find_col([r"^estado\s*final$"], df.columns)
col_programa     = find_col([r"^programa$"], df.columns)
col_facultad     = find_col([r"^facultad$"], df.columns)
col_plazo        = find_col([r"^plazo$"], df.columns)
col_ingresos     = find_col([r"^total\s*ingresos$","ingresos"], df.columns)
col_egresos      = find_col([r"^total\s*egresos$","egresos"], df.columns)

def to_number(s):
    return pd.to_numeric(
        s.astype(str)
         .str.replace(r"[^\d,\.-]","", regex=True)
         .str.replace(".","", regex=False)
         .str.replace(",",".", regex=False),
        errors="coerce"
    )

df["valor_financiado"] = pd.NA
if col_valor_liq:  df["valor_financiado"] = to_number(df[col_valor_liq])
elif col_valor_fact: df["valor_financiado"] = to_number(df[col_valor_fact])
elif col_valor_fin:  df["valor_financiado"] = to_number(df[col_valor_fin])

if col_valor_mora:  df["valor_mora"]      = to_number(df[col_valor_mora])
if col_media_score: df["media_score"]     = to_number(df[col_media_score])
if col_plazo:       df["plazo"]           = pd.to_numeric(df[col_plazo], errors="coerce").astype("Int64")
if col_ingresos:    df["total_ingresos"]  = to_number(df[col_ingresos])
if col_egresos:     df["total_egresos"]   = to_number(df[col_egresos])

def parse_mes(m):
    if pd.isna(m): return np.nan
    s=str(m).strip().lower()
    mapa = {"ene":1,"enero":1,"feb":2,"febrero":2,"mar":3,"marzo":3,"abr":4,"abril":4,"may":5,"mayo":5,
            "jun":6,"junio":6,"jul":7,"julio":7,"ago":8,"agosto":8,"sep":9,"sept":9,"septiembre":9,
            "oct":10,"octubre":10,"nov":11,"noviembre":11,"dic":12,"diciembre":12}
    if s.isdigit(): 
        n=int(float(s)); 
        return n if 1<=n<=12 else np.nan
    return mapa.get(s, np.nan)
if col_mes and col_anio:
    mes_num = pd.Series(df[col_mes]).map(parse_mes)
    anio_num = pd.to_numeric(df[col_anio], errors="coerce").astype("Int64")
    df["periodo"] = (anio_num*100 + mes_num.fillna(0).astype("Int64")).where(anio_num.notna() & mes_num.notna())

if col_estado_final: df["estado_final"] = df[col_estado_final].astype("string").str.strip()
if col_programa:     df["programa"]     = df[col_programa].astype("string").str.strip()
if col_facultad:     df["facultad"]     = df[col_facultad].astype("string").str.strip()

# mora_flag: por “Mora” (si existe) o por valor_mora>0
def infer_mora_flag(row):
    if col_mora_text and pd.notna(row.get(col_mora_text)):
        s=str(row[col_mora_text]).strip().lower()
        if s in {"1","si","sí","true","mora","en mora","con mora"}: return 1
        if s in {"0","no","false","sin mora","al dia","al día"}:     return 0
    vm = row.get("valor_mora")
    return 1 if (pd.notna(vm) and vm>0) else 0
df["mora_flag"] = df.apply(infer_mora_flag, axis=1).astype("int8")

if {"total_ingresos","total_egresos"}.issubset(df.columns):
    df["ratio_endeudamiento"] = (df["total_egresos"]/df["total_ingresos"]).replace([np.inf,-np.inf], np.nan)

# Tablas del informe
present = [c for c in ["valor_financiado","valor_mora","total_ingresos","total_egresos","ratio_endeudamiento",
                       "plazo","media_score","estado_final","periodo","programa","facultad","mora_flag"] if c in df.columns]

monto_total = float(df["valor_financiado"].sum()) if "valor_financiado" in present else np.nan
mora_total  = float(df["valor_mora"].sum()) if "valor_mora" in present else np.nan
pct_reg_mora = round(df["mora_flag"].mean()*100,2) if "mora_flag" in present else np.nan

tabla_ejecutiva = pd.DataFrame({
    "observaciones":[len(df)],
    "periodo_min":[int(df["periodo"].min()) if "periodo" in present and df["periodo"].notna().any() else np.nan],
    "periodo_max":[int(df["periodo"].max()) if "periodo" in present and df["periodo"].notna().any() else np.nan],
    "monto_total_financiado":[monto_total],
    "mora_total":[mora_total],
    "ticket_medio":[float(df["valor_financiado"].mean()) if "valor_financiado" in present else np.nan],
    "ingresos_medios":[float(df["total_ingresos"].mean()) if "total_ingresos" in present else np.nan],
    "egresos_medios":[float(df["total_egresos"].mean()) if "total_egresos" in present else np.nan],
    "ratio_endeudamiento_mediana":[float(df["ratio_endeudamiento"].median()) if "ratio_endeudamiento" in present else np.nan],
    "porc_registros_con_mora":[pct_reg_mora],
})
display(tabla_ejecutiva.style.format({
    "monto_total_financiado":"{:,.0f}","mora_total":"{:,.0f}","ticket_medio":"{:,.0f}",
    "ingresos_medios":"{:,.0f}","egresos_medios":"{:,.0f}","ratio_endeudamiento_mediana":"{:.2f}",
    "porc_registros_con_mora":"{:.2f}"
}))

qc = pd.DataFrame({
    "variable": present,
    "n_no_nulos": [df[v].notna().sum() for v in present],
    "porc_faltantes": [round(df[v].isna().mean()*100,2) for v in present],
    "cardinalidad": [df[v].nunique(dropna=True) for v in present],
    "dtype": [str(df[v].dtype) for v in present],
}).sort_values("variable").reset_index(drop=True)
display(qc)

def tabla_pct(series, top=None):
    s=series.dropna()
    if s.empty: return pd.DataFrame(columns=["categoria","pct"])
    t=(s.value_counts(normalize=True)*100).sort_values(ascending=False)
    if top: t=t.head(top)
    return t.round(2).rename_axis("categoria").reset_index(name="pct")

tbl_estado = tabla_pct(df["estado_final"]) if "estado_final" in present else pd.DataFrame()
tbl_prog   = tabla_pct(df["programa"], top=15) if "programa" in present else pd.DataFrame()
tbl_fac    = tabla_pct(df["facultad"], top=10) if "facultad" in present else pd.DataFrame()
tbl_per    = tabla_pct(df["periodo"]) if "periodo" in present else pd.DataFrame()
tbl_mora   = tabla_pct(df["mora_flag"].map({0:"no mora",1:"mora"})) if "mora_flag" in present else pd.DataFrame()

print("Composición por estado_final (%)"); display(tbl_estado)
print("Top programas (%)");                 display(tbl_prog)
print("Top facultades (%)");                display(tbl_fac)
print("Distribución por periodo (%)");      display(tbl_per.head(15))
print("Mora vs no mora (%)");               display(tbl_mora)

# Gráficos
if "estado_final" in present: bar_pct(df["estado_final"], "Composición por estado_final", "estado_final", rot=45)
if "programa" in present:     bar_pct(df["programa"], "Top programas por participación", "programa", top=15, rot=45)
if "facultad" in present:     bar_pct(df["facultad"], "Top facultades por participación", "facultad", top=10, rot=45)
if "periodo" in present:      bar_pct(df["periodo"], "Distribución por periodo", "periodo", rot=45)

if "valor_financiado" in present: hist_nice(df["valor_financiado"].clip(lower=0), "Distribución de valor_financiado", "valor_financiado")
if "valor_mora" in present:       hist_nice(df.loc[df["valor_mora"]>0,"valor_mora"], "Distribución de valor_mora (>0)", "valor_mora")
if "total_ingresos" in present:   hist_nice(df["total_ingresos"].clip(lower=0), "Distribución de total_ingresos", "total_ingresos")
if "total_egresos" in present:    hist_nice(df["total_egresos"].clip(lower=0), "Distribución de total_egresos", "total_egresos")

if {"valor_financiado","estado_final"}.issubset(present):
    box_by_cat(df["valor_financiado"], df["estado_final"],
               "valor_financiado por estado_final (top 12)", "estado_final", "valor_financiado")

# Mix de mora por periodo 
def _to_number(s):
    return pd.to_numeric(
        s.astype(str)
         .str.replace(r"[^\d,\.-]", "", regex=True) 
         .str.replace(".", "", regex=False)
         .str.replace(",", ".", regex=False),
        errors="coerce"
    )

col_valor_mora = None
for cand in ["Valor Mora", "valor_mora", "VALOR MORA"]:
    if cand in df.columns:
        col_valor_mora = cand
        break

if col_valor_mora is not None:
    df["_valor_mora_num"] = _to_number(df[col_valor_mora]).fillna(0)
    df["mora_flag"] = (df["_valor_mora_num"] > 0).astype("int8")
else:
    print("No se encontró la columna 'Valor Mora'. No se actualizará 'mora_flag'.")

if {"periodo", "mora_flag"}.issubset(df.columns):
    pv = pd.crosstab(
        df["periodo"],
        df["mora_flag"].map({0: "no mora", 1: "mora"}),
        normalize="index"
    ) * 100

    if not pv.empty:
        pv = pv.sort_index()
        fig, ax = plt.subplots(figsize=(12.5, 4.5))
        left = np.zeros(len(pv))
        for c in pv.columns:
            vals = pv[c].values
            ax.bar(
                pv.index.astype(str), vals, bottom=left, label=str(c),
                edgecolor="white", linewidth=0.5
            )
            left += vals

        ax.set_title("Mix de mora por periodo (% por periodo)")
        ax.set_xlabel("periodo")
        ax.set_ylabel("% dentro de periodo")
        ax.yaxis.set_major_formatter(PercentFormatter(100))
        ax.legend(title="mora_flag", ncols=min(len(pv.columns), 8),
                  bbox_to_anchor=(1.02, 1), loc="upper left")

        plt.xticks(rotation=45, ha="right")
        ax.margins(x=0.03)
        plt.subplots_adjust(left=0.07, right=0.92, bottom=0.25, top=0.90)
        plt.show()
else:
    print("Faltan columnas 'periodo' o 'mora_flag' para graficar.")



```
